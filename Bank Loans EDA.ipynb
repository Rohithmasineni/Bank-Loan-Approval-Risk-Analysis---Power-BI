#### Univariate Analysis

```python
for col in num_cols:
    plt.figure(figsize=(8,3))
    sns.histplot(x=data[col], kde=True)
    plt.show()
```

```python
cols = ['Gender', 'Banking Relationship', 'Risk Weighting', 'Fee Structure', 'Nationality', 'Loyalty Classification', 'Investment Advisor']
```

```python
for col in cols:
    fig,axes = plt.subplots(1,2,figsize=(8,3), layout='constrained')
    plt.suptitle('Categorical Data Distributions')

    axes[0].set_title('Bar Chart')
    sns.countplot(x=col, data=data, ax=axes[0])

    axes[1].set_title('Pie Chart')
    vc_data = data[col].value_counts()
    axes[1].pie(vc_data, labels = vc_data, autopct = '%1.1f%%')
    
    fig.show()
```

```python

# bins and labels
bins = [17, 27, 37, 47, 57, 67, 77, 85]
labels = ["17-26", "27-36", "37-46", "47-56", "57-66", "67-77", "78-85"]

# Plotting directly using temporary binned series

plt.figure(figsize=(10, 5))
ax = sns.barplot(
    x=pd.cut(data["Age"], bins=bins, labels=labels, right=True),
    y=data["Bank Loans"],
    estimator=np.mean,
    errorbar=None
)
for bar in ax.containers:
    ax.bar_label(bar,fmt='%0.2f')

plt.xticks(rotation=45)
plt.title("Average Bank Loans by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Bank Loan")
plt.tight_layout()
plt.show()

# Do younger customers take more loans than older ones...?
```
